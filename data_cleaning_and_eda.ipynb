游녤 Performed Data Cleaning & EDA to handle missing values, remove duplicates, fix data types, and analyze patterns, trends, and outliers for reliable insights

```python
df.isnull()   #checks the dataset for missing values and marks them as True, helping identify data quality issues before analysis.
```

```python
df.isnull().sum() #counts the total number of missing values in each column, helping assess data completeness quickly.
```

```python
df.dropna()  #removes rows containing missing (NaN) values to ensure clean and reliable data for analysis.
```

```python
df.dropna(axis=1) #removes columns that contain missing (NaN) values to keep only complete features for analysis
```

DATE COLUMNS FIX

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df.info()
```

Feature Engineering (I created new features like Year, Month, and Profit Margin for time-based analysis.)

```python
df['Year'] = df['order_date'].dt.year
df['Month'] = df['order_date'].dt.month
df['Profit_Margin'] = (df['profit'] / df['sales']) * 100
```

Exploratory Data Analysis (EDA)

```python
df[['sales','profit','quantity']].describe() #
```

Sales by Region (calculates total sales per region and sorts them from highest to lowest to identify top-performing regions.)

```python
df.groupby('region')['sales'].sum().sort_values(ascending=False)
```

Profit by Category (calculates total profit for each product category to compare category-wise performance.)

```python
df.groupby("category")['profit'].sum() 
```

insights -> Technology generates the highest profit, followed by Office Supplies, while Furniture contributes the least profit, indicating Technology is the most profitable category and Furniture needs improvement.

Visualization (Matplotlib + Seaborn)

Sales by Category

```python
sns.barplot(x='category', y='sales', data=df)
plt.title("sales by category")
plt.show()
```

```python
#Technology category generates highest sales
```

Monthly Sales Trend

```python
monthly_sales = df.groupby('Month')['sales'].sum()

plt.plot(monthly_sales.index, monthly_sales.values)
plt.xlabel("Month")
plt.ylabel("sales")
plt.title("Monthly Sales Trend")
plt.show()
```

Discount vs Profit

```python
sns.scatterplot(x='discount', y='profit', data=df)
plt.title("Discount vs Profit")
plt.show()
```

#Higher discounts often result in negative profit

Which category is selling more but still making loss?

```python
cat_sales_profit = df.groupby('category')[['sales','profit']].sum().reset_index()

sns.barplot(x='category', y='sales', data=cat_sales_profit)
plt.title("category-wise sales")
plt.show()

sns.barplot(x='category', y='profit', data=cat_sales_profit)
plt.title("category-wise profit")
plt.show()
```

```python
#Furniture has good sales but negative profit  pricing/discount issue.
```

Which region is profitable and which is risky?

```python
region_profit = df.groupby('region', as_index=False)['profit'].sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='region', y='profit', data=region_profit)
plt.title("Region-wise Profit")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Some regions have strong sales but weak profitability.
```

Which customers generate most revenue?

```python
top_customers = df.groupby('customer_name')['sales'].sum().sort_values(ascending=False).head(10)

sns.barplot(x=top_customers.values, y=top_customers.index)
plt.title("Top 10 Customers by Sales")
plt.show()
```

```python
#Few customers contribute large portion of revenue.
```

Profit margin comparison by category

```python
sns.boxplot(x='category', y='Profit_Margin', data=df)
plt.title("Profit Margin Distribution by Category")
plt.show()
```

```python
#Technology has stable margins, Furniture highly volatile.
```
